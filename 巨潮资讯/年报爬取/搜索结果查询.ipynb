```python
import requests,tqdm,pandas
```

```python
def search(name):
    url = "http://www.cninfo.com.cn/new/fulltextSearch/full"
    params = {
    "searchkey": name,
    "sdate": "",
    "edate": "",
    "isfulltext": "true",
    "sortName": "nothing",
    "sortType": "desc",
    "pageNum": "1",
    "type": ""
    }
    return  requests.get(url,  params=params, verify=False)
```

```python
js = search("山东卓锐石化科技有限公司").json()
company_id = [{k:js['announcements'][i][k] for k in ['secName','announcementTitle','secCode', 'orgId']} for i in range(len(js['announcements']))]
company_id
```

```python
with open('/Users/nanapower/学习资料/python_learnning/模块与项目/AI/图灵AI/巨潮资讯/年报爬取/企业全称名单.txt','r') as f:
    company_list = f.read().split('\n')
```

```python
pandas.DataFrame([{'secName': '无',
  'announcementTitle': '无',
  'secCode': '无',
  'orgId': '无'}])
```

```python
result_dict = {}
for i in tqdm.tqdm(company_list[:1000]):
    js = search(i).json()
    if js['totalpages']==0:
        result_dict[i]=pandas.DataFrame([{'secName': '无',
  'announcementTitle': '无',
  'secCode': '无',
  'orgId': '无'}])
        continue
    company_id = [{k:js['announcements'][i][k] for k in ['secName','announcementTitle','secCode', 'orgId']} for i in range(len(js['announcements']))]
    result_dict[i]=pandas.DataFrame(company_id[:5])
```

```python
df = pandas.concat(result_dict)
df
```
