```python
import requests
from concurrent.futures.thread import ThreadPoolExecutor
```

```python
with open('/Users/nanapower/学习资料/python_learnning/模块与项目/AI/图灵AI/巨潮资讯/年报爬取/企业简称名单.txt','r') as f:
    company_list = f.readlines()
```

```python
company_list = [i.strip() for i in company_list]
company_list
```

```python
# 搜索功能
def search(name):
    # 向搜索接口发起请求,获取数据
    params = {
        'keyWord': name,
        'maxNum': 5
    }
    r = requests.post('http://www.cninfo.com.cn/new/information/topSearch/query', params=params)
    # 过滤A股
    data = []
    for i in r.json():
        data.append(i)

    # 判断是否无搜索结果，如果无搜索结果就重新循环
    if len(data) == 0:
        f.write(str({name:'None'})+',')
        return
    f.write(str({name:data})+',')
```

```python
f = open('企业id获取结果.txt','a')
with ThreadPoolExecutor(50) as pool:
    result = pool.map(search, company_list)
f.close()
```
