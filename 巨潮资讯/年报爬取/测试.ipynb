```python

# 功能一
def search():
    while True:
        # 输入关键词
        keyword = input('请输入搜索关键词：')
        # 向搜索接口发起请求,获取数据
        params = {
            'keyWord': keyword,
            'maxNum': 10
        }
        r = requests.post('http://www.cninfo.com.cn/new/information/topSearch/query', params=params)
        # 过滤A股
        data = []
        for i in r.json():
            data.append(i)

        # 判断是否无搜索结果，如果无搜索结果就重新循环
        if len(data) == 0:
            return None
        return data
```

```python

# 功能二
def select(code, orgid):

    # 获取报告类型
    category="category_ndbg_szsh;"

    # 获取时间
    start = '2013-01-01'
    end = '2023-04-21'

    # 根据股票代码的头文字，判断股票交易所信息
    if str(code)[0] == '6':
        column = 'sse'
        plate = 'sh'
    else:
        column = 'szse'
        plate = 'sz'

    # 设置初始页码
    page_num = 1
    # 设置初始列表存储筛选结果
    pdf_list = []
    while True:
        # 设置报告筛选参数
        params = {
            'stock': '{},{}'.format(code, orgid),
            'tabName': 'fulltext',
            'pageSize': '5',
            'pageNum': str(page_num),
            'category': '',#category,
            'seDate': '{}~{}'.format(start, end),
            'column': 'third',
            'plate': 'neeq',
            'searchkey': '',
            'secid': '',
            'sortName': '',
            'sortType': '',
            'isHLtitle': ''
        }

        # 发起报告搜索请求
        r = requests.post('http://www.cninfo.com.cn/new/hisAnnouncement/query', params=params)
        r_json = r.json()

        # 判断是否搜索失败、或者无搜索结果，如果无结果则结束
        if r_json['announcements'] == None or len(r_json['announcements']) == 0:
            print('无搜索结果！')
            break

        # 遍历搜索结果
        for i in r_json['announcements']:
            pdf_list.append([i['announcementTitle'], i['adjunctUrl']])
        print(type(r_json['hasMore']))
        # 判断是否还有下一页数据，没有的话就结束循环
        if r_json['hasMore'] != True:
            break

        # 如果有的话，让页数加一，开始下一轮循环
        page_num += 1
        
    return pdf_list
```

```python
# 功能三
def download(pdf_list):
    # 循环遍历筛选结果
    for item in pdf_list:

        # 拼接完整 url
        pdf_r = requests.get('http://static.cninfo.com.cn/' + item[1])

        # 拼接文件后缀
        file_path = './年报pdf/'+item[0] + '.pdf'

        # 将报告内容写入文件，保存文件
        with open(file_path, 'wb') as f:
            f.write(pdf_r.content)
        
        # 打印下载提示
        print('已完成{}的下载'.format(item[0]))

```

```python
# 功能四
def pdf2txt(pdf_file):
    with open('./年报txt/'+pdf_file,'w') as f:
        # 创建一个PDF Reader对象
        pdf_reader = PyPDF2.PdfFileReader('./年报pdf/'+pdf_file)
        # 遍历每页并提取文本
        for page_num in range(pdf_reader.numPages):
            page = pdf_reader.getPage(page_num)
            f.write(page.extractText())
```

```python
# 功能五
def Word_frequency_analysis(txt_file,type):
    with open(txt_file, 'r', encoding='utf-8') as f:
        text = f.read()
     # 精确模式分词
    if type == '精确':
        words = list(jieba.lcut(text))
    if type == '全模式':
        words = list(jieba.lcut(text,cut_all=True))
    if type == '搜索引擎':
        words = list(jieba.lcut_for_search(text))
    # 统计词频
    word_counts = Counter(words)
    # 输出结果
    for word, count in word_counts.most_common():
        print(word, count)
```
