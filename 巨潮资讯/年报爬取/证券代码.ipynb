```python
# 功能一
def search(num):
    while True:
        # 输入关键词
        # 向搜索接口发起请求,获取数据
        params = {
            'keyWord': num,
            'maxNum': 10
        }
        r = requests.post('http://www.cninfo.com.cn/new/information/topSearch/query', params=params)
        # 过滤A股
        data = []
        for i in r.json():
            data.append(i)
            break
        
```
