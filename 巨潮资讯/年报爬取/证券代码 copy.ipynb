```python
# 功能一
def search(num):
    length = len(str(num))
    if length<6:
        num = '0'*(6-length)+str(num)
    print(num)
    try:
        # 输入关键词
        # 向搜索接口发起请求,获取数据
        params = {
            'keyWord': num,
            'maxNum': 10
        }
        r = requests.post('http://www.cninfo.com.cn/new/information/topSearch/query', params=params)
        # 过滤A股
        for i in r.json():
            data = i
            break
        return [data['code'],data['zwjc']]
    except:
        return ['无','无']
        
```
