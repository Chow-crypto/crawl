```python
import jieba,os
from collections import Counter
```

```python
# 功能五
def Word_frequency_analysis(txt_file,type):
    with open(txt_file, 'r', encoding='utf-8') as f:
        text = f.read()
     # 精确模式分词
    if type == '精确':
        words = list(jieba.lcut(text))
    if type == '全模式':
        words = list(jieba.lcut(text,cut_all=True))
    if type == '搜索引擎':
        words = list(jieba.lcut_for_search(text))
    # 统计词频
    word_counts = Counter(words)
    # 输出结果
    for word, count in word_counts.most_common():
        print(word, count)
```
